```python
# The different rock types
classes = np.array(['ML', 'MS','BL','GN','OL'])

fig, ax = plt.subplots(2,5,figsize=(10,4))

# ML rock type
ax[0,0].imshow(Cuttings(df_test.loc[0,'path'][:24],df_test.loc[0,'path'][24:]).load_picture(),cmap='gray')
ax[1,0].imshow(Cuttings(df_test.loc[25,'path'][:24],df_test.loc[25,'path'][24:]).load_picture(),cmap='gray')
ax[0,0].axis('off')
ax[1,0].axis('off')
ax[0,0].title.set_text(classes[df_test.loc[0,'rock_type']])
ax[1,0].title.set_text(classes[df_test.loc[25,'rock_type']])

# MS rock type
ax[0,1].imshow(Cuttings(df_test.loc[200,'path'][:24],df_test.loc[200,'path'][24:]).load_picture(),cmap='gray')
ax[1,1].imshow(Cuttings(df_test.loc[250,'path'][:24],df_test.loc[250,'path'][24:]).load_picture(),cmap='gray')
ax[0,1].axis('off')
ax[1,1].axis('off')
ax[0,1].title.set_text(classes[df_test.loc[200,'rock_type']])
ax[1,1].title.set_text(classes[df_test.loc[250,'rock_type']])

# BL rock type
ax[0,2].imshow(Cuttings(df_test.loc[400,'path'][:24],df_test.loc[400,'path'][24:]).load_picture(),cmap='gray')
ax[1,2].imshow(Cuttings(df_test.loc[410,'path'][:24],df_test.loc[410,'path'][24:]).load_picture(),cmap='gray')
ax[0,2].axis('off')
ax[1,2].axis('off')
ax[0,2].title.set_text(classes[df_test.loc[400,'rock_type']])
ax[1,2].title.set_text(classes[df_test.loc[410,'rock_type']])

# GN rock type
ax[0,3].imshow(Cuttings(df_test.loc[615,'path'][:24],df_test.loc[615,'path'][24:]).load_picture(),cmap='gray')
ax[1,3].imshow(Cuttings(df_test.loc[640,'path'][:24],df_test.loc[640,'path'][24:]).load_picture(),cmap='gray')
ax[0,3].axis('off')
ax[1,3].axis('off')
ax[0,3].title.set_text(classes[df_test.loc[615,'rock_type']])
ax[1,3].title.set_text(classes[df_test.loc[640,'rock_type']])

# OL rock type
ax[0,4].imshow(Cuttings(df_test.loc[800,'path'][:24],df_test.loc[800,'path'][24:]).load_picture(),cmap='gray')
ax[1,4].imshow(Cuttings(df_test.loc[820,'path'][:24],df_test.loc[820,'path'][24:]).load_picture(),cmap='gray')
ax[0,4].axis('off')
ax[1,4].axis('off')
ax[0,4].title.set_text(classes[df_test.loc[800,'rock_type']])
ax[1,4].title.set_text(classes[df_test.loc[820,'rock_type']])

# Save figure
plt.tight_layout()
plt.savefig('rock_type_problem_name.png',bbox_inches='tight')
```
